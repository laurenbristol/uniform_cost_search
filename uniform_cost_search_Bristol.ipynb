```python
from queue import PriorityQueue


def  uniform_cost_search(start, goal, graph, cost):
    answer = {node: {'cost': float('inf'), 'path': []} for node in goal}

    frontier = PriorityQueue()
    visited = []

    #make priority queue with priority sorted by cost to get to the start node (zero) and the number of the node itself (start) and initialize path to start node as itself

    frontier.put((0, start, [start]))
    
    # x = 1
    while not frontier.empty():
        
        currentCost, currentNode, currentPath= frontier.get()

        # print("before: current path is... " + str(currentPath))
        
        if currentNode not in visited: 
            for child in graph[currentNode]:
                #only expand current node
                #add updated cost, the node we expanded to, and add the node we expanded to to the path that it took to get to the node we are expanding (current node)
                # print("adding " + str((currentNode, child)) + ", " + str(cost[(currentNode, child)]) + ", " + str(currentPath+[child]) + " to frontier")  
                frontier.put((cost[(currentNode, child)]+currentCost, child, currentPath+[child]))

            
            if currentNode in goal:
                answer[currentNode] = {'cost': currentCost, 'path': currentPath}
                
            visited.append(currentNode)
            
            # print("current cost: " + str(currentCost))
            # print("current node: " + str(currentNode))
            # print("current path: " + str(currentPath))
            # print("visited: " + str(visited))
            # print("------------ROUND " + str(x) + " DONE-------------------")
            # x+=1

    return answer
```
